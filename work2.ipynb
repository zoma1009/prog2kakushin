```python
import numpy as np
import ipywidgets as wid
from google.colab import output
from time import sleep
run=0
ehpo=100
ehp=100
hhp=100
hhpo=100
hatk=10
hdef=5
eatk=10
edef=5
com=0

battle=['たたかう','かいふく']
def main():
  bu1=wid.Button(description=battle[0])
  bu1.on_click(fight)
  bu2=wid.Button(description=battle[1])
  bu2.on_click(heal)
  return wid.VBox([wid.HBox([bu1,bu2])])

def enemy(i,n):
  a=i*10/n
  print("\r"+'enemy HP:'+'■'*int(a)+'□'*(10-int(a)),end="")

def human(i,n):
  print("\r"+'human HP:'+str(i)+'/'+str(n),end="")

def fight(n):
    global ehp,hatk,edef
    ehp=ehp-(hatk-edef)
    print("\r"+'プレイヤー　の　こうげき',end="")
    sleep(1)
    enemy(ehp,ehpo)
    sleep(1)
    enemy_action()

def heal(self):
  global hhp
  n=np.random.randint(1,11)
  hhp=hhp+n
  if hhp>100:
    n=n-(hhp-100)
    hhp=100
  print("\r"+'プレイヤー　は　'+str(n)+' かいふくした',end="")
  sleep(1)
  human(hhp,hhpo)
  sleep(1)
  enemy_action()

def enemy_action():
  ac=np.random.randint(0,4)
  print(ac)
  if ac==0:
    global hhp,eatk,hdef
    hhp=hhp-(eatk-hdef)
    print("\r"+"あいて　の　こうげき",end="")
    sleep(1)
    human(hhp,hhpo)
  if ac==1:
    hhp=hhp-10
    print("\r"+'あいて　は　ほのお　を　ふいた',end="")
    sleep(1)
    human(hhp,hhpo)
  if ac==2:
    global ehp
    ehp=ehp+5
    if ehp>100:
      ehp=100
    print("\r"+'あいて　は　かいふく　した',end="")
    sleep(1)
    human(hhp,hhpo)
  if ac==3:
    print("\r"+'あいては　なにも　しなかった',end="")
    sleep(1)
    human(hhp,hhpo)

main()
```
