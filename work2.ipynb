```python
import numpy as np
import ipywidgets as wid
from google.colab import output
from time import sleep
ehpo=200
ehp=200
hhp=100
hhpo=100
hatk=10
hdef=5
eatk=15
edef=5

battle=['たたかう','かいふく']
def main():
  bu1=wid.Button(description=battle[0])
  bu1.on_click(fight)
  bu2=wid.Button(description=battle[1])
  bu2.on_click(heal)
  return wid.VBox([wid.HBox([bu1,bu2])])

def enemy(i,n):
  a=(i*10/n)+1
  if i==0:
    a=0
  print("\r"+'enemy HP:'+'■'*int(a)+'□'*(11-int(a)),end="")

def human(i,n):
  if i<0:
    i=0
  print("\r"+'human HP:'+str(i)+'/'+str(n),end="")

def fight(n):
    global ehp,hatk,edef
    crt=np.random.randint(1,33)
    if crt==1:
      crt=2
    else:crt=1
    dog=+np.random.randint(1,65)
    print("\r"+'プレイヤー　の　こうげき',end="")
    sleep(1)
    if dog==1:
      sleep(1)
      print("\r"+'あいて　は　すばやく　みを　かわした！',end='')
    elif crt==2:
      sleep(1)
      print("\r"+'かいしんのいちげき！',end='')
      dmg=(hatk-edef+np.random.randint(-4,6))*crt
      ehp=ehp-dmg
      print("\r"+'あいて　に　'+str(dmg)+'　の　ダメージ！',end='')
    else:
      dmg=(hatk-edef+np.random.randint(-4,6))
      ehp=ehp-dmg
      print("\r"+'あいて　に　'+str(dmg)+'　の　ダメージ！',end='')
    sleep(1)
    enemy(ehp,ehpo)
    sleep(1)
    judge('e')
    if ehp>0:
      enemy_action()

def heal(self):
  global hhp
  n=np.random.randint(5,21)
  hhp=hhp+n
  if hhp>100:
    n=n-(hhp-100)
    hhp=100
  print("\r"+'プレイヤー　は　'+str(n)+' かいふくした',end="")
  sleep(1)
  human(hhp,hhpo)
  sleep(1)
  enemy_action()

def enemy_action():
  ac=np.random.randint(0,10)
  if ac==0 or ac==1 or ac==2 or ac==3:
    global hhp,eatk,hdef
    dog=+np.random.randint(1,65)
    print("\r"+"あいて　の　こうげき",end="")
    sleep(1)
    if dog==1:
      print("\r"+'プレイヤー　は　すばやく　みを　かわした！',end='')
    else:
      dmg=(eatk-hdef+np.random.randint(-9,6))
      hhp=hhp-dmg
      print("\r"+'プレイヤー　に　'+str(dmg)+'　の　ダメージ！',end='')
    sleep(1)
    human(hhp,hhpo)
    sleep(1)
    judge('h')
  if ac==4 or ac==5:
    hhp=hhp-10
    print("\r"+'あいて　は　ほのお　を　ふいた',end="")
    sleep(1)
    print("\r"+'プレイヤー　に　'+str(10)+'　の　ダメージ！',end='')
    sleep(1)
    human(hhp,hhpo)
    sleep(1)
    judge('h')
  if ac==6:
    global ehp
    ehp=ehp+5
    if ehp>100:
      ehp=100
    print("\r"+'あいて　は　かいふく　した',end="")
    sleep(1)
    human(hhp,hhpo)
  if ac==7 or ac==8 or ac==9:
    print("\r"+'あいては　なにも　しなかった',end="")
    sleep(1)
    human(hhp,hhpo)
    sleep(1)
    judge('h')

def judge(who):
  global hhp,ehp
  if who=='h' and hhp<=0:
    output.clear()
    print('やられてしまった')
  if who=='e' and ehp<=0:
    output.clear()
    print('てきをやっつけた')

main()
```
